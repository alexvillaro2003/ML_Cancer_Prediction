```python
# Importacion de Librer√≠as:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

---
**<h3 style="text-align:center;">üîß: <u>Importaci√≥n</u> + <u>Limpieza</u></h3>**

```python
# Importamos 'The_Cancer_data_1500_V2.csv'
df = pd.read_csv('../data/raw/The_Cancer_data_1500_V2.csv')


# Dummies para la columna 'GeneticRisk'
df_riesgos = pd.get_dummies(df['GeneticRisk'], prefix='Riesgo')

df['Riesgo_0'] = df_riesgos['Riesgo_0'].astype(int)
df['Riesgo_1'] = df_riesgos['Riesgo_1'].astype(int)
df['Riesgo_2'] = df_riesgos['Riesgo_2'].astype(int)

df.head()
```

---
**<h3 style="text-align:center;">‚¨áÔ∏è Guardar: '<u>processed.csv</u>' </h3>**

```python
df_processed = df
df_processed.to_csv('../data/processed.csv')
```

---
**<h3 style="text-align:center;">‚ûó Separaci√≥n: '<u>Data</u>'‚≠§   '<u>Target</u>' </h3>**

```python
X = df.drop(['Diagnosis', 'GeneticRisk'], axis=1)
y = df['Diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

---
**<h3 style="text-align:center;">‚¨áÔ∏è Guardar: '<u>train.csv</u>' </h3>**

```python
df_train_csv = pd.DataFrame(X_train)
df_train_csv['Diagnosis'] = y_train
df_train_csv.to_csv('../data/train.csv')

df_train_csv.head()
```

---
**<h3 style="text-align:center;">‚¨áÔ∏è Guardar: '<u>test.csv</u>' </h3>**

```python
df_test_csv = pd.DataFrame(X_test)
df_test_csv['Diagnosis'] = y_test
df_test_csv.to_csv('../data/test.csv')

df_test_csv.head()
```
