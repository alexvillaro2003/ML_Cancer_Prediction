___
**<h3 style="text-align:center;">✏️ <u>Descripción de variables</u> ✏️</h3>**

| ***<u>Variable</u>***               | ***<u>Descripción</u>***                                                                                                                | ***<u>Tipo</u>***   |
|-------------------------------------|-----------------------------------------------------------------------------------------------------------------------------------------|----------------------|
| **Country**                         | El país al que pertenece el registro.                                                                                                   | Categórica           |
| **Year**                            | Año del registro (de 2000 a 2015).                                                                                                      | Numérica             |
| **Status**                          | Estatus de desarrollo del país (Desarrollado o en Desarrollo).                                                                          | Categórica           |
| **Life expectancy**                 | Esperanza de vida en años.                                                                                                              | Numérica             |
| **Adult Mortality**                 | Tasa de mortalidad adulta (probabilidad de morir entre los 15 y 60 años por cada 1000 personas).                                       | Numérica             |
| **infant deaths**                   | Número de muertes infantiles por cada 1000 personas.                                                                                    | Numérica             |
| **Alcohol**                         | Consumo per cápita de alcohol (en litros de alcohol puro) para personas de 15 años o más.                                              | Numérica             |
| **percentage expenditure**          | Gasto en salud como porcentaje del PIB per cápita.                                                                                      | Numérica             |
| **Hepatitis B**                     | Cobertura de inmunización contra la Hepatitis B en niños de 1 año (%).                                                                  | Numérica             |
| **Measles**                         | Número de casos reportados de sarampión por cada 1000 personas.                                                                         | Numérica             |
| **BMI**                              | Índice de masa corporal promedio de la población.                                                                                       | Numérica             |
| **under-five deaths**               | Número de muertes en menores de cinco años por cada 1000 personas.                                                                      | Numérica             |
| **Polio**                           | Cobertura de inmunización contra la polio en niños de 1 año (%).                                                                        | Numérica             |
| **Total expenditure**               | Gasto general del gobierno en salud como porcentaje del gasto total del gobierno.                                                       | Numérica             |
| **Diphtheria**                      | Cobertura de inmunización contra la difteria en niños de 1 año (%).                                                                     | Numérica             |
| **HIV/AIDS**                        | Número de muertes relacionadas con el VIH/SIDA por cada 1000 nacimientos vivos.                                                         | Numérica             |
| **GDP**                             | Producto Interno Bruto (PIB) per cápita en dólares americanos (USD).                                                                    | Numérica             |
| **Population**                      | Población total del país.                                                                                                               | Numérica             |
| **thinness 1-19 years**             | Prevalencia de delgadez entre adolescentes de 10 a 19 años (%).                                                                         | Numérica             |
| **thinness 5-9 years**              | Prevalencia de delgadez entre niños de 5 a 9 años (%).                                                                                  | Numérica             |
| **Income composition of resources** | Índice de desarrollo humano en términos de composición de ingresos (de 0 a 1).                                                          | Numérica             |
| **Schooling**                       | Años promedio de escolarización.                                                                                                        | Numérica             |

```python
# Distribución Variable Objetivo:
plt.figure(figsize=(16, 6))
sns.histplot(df['Life expectancy '], bins=50)
plt.title("Distribución: 'Life expectancy'")
plt.xlabel("Esperanza de vida")
plt.ylabel("Frecuencia");
```

```python
# Outliers Variable Objetivo:
plt.figure(figsize=(16, 6))
sns.boxplot(x=df['Life expectancy '])
plt.title("Outliers 'Life expectancy'");
```

```python
plt.figure(figsize=(16,16))
sns.heatmap(df.corr(numeric_only=True), annot = True, fmt = '.2f', vmin = -1, cmap = 'coolwarm');
```

```python
sns.heatmap(df[['Adult Mortality', 'Alcohol', 'percentage expenditure', ' BMI ', 'Polio', 'Diphtheria ', 'GDP', 'Income composition of resources', 'Schooling', 'Life expectancy ']].corr(), annot= True, vmin=-1, cmap='coolwarm');
```
