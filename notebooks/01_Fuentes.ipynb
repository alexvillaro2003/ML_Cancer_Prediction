**<h3 style="text-align:center;">‚úèÔ∏è <u>Descripci√≥n de variables</u> ‚úèÔ∏è</h3>**

| ***<u>Variable</u>***               | ***<u>Descripci√≥n</u>***                                                                                                                | ***<u>Tipo</u>***   |
|-------------------------------------|-----------------------------------------------------------------------------------------------------------------------------------------|----------------------|
| **Country**                         | El pa√≠s al que pertenece el registro.                                                                                                   | Categ√≥rica           |
| **Year**                            | A√±o del registro (de 2000 a 2015).                                                                                                      | Num√©rica             |
| **Status**                          | Estatus de desarrollo del pa√≠s (Desarrollado o en Desarrollo).                                                                          | Categ√≥rica           |
| **Life expectancy**                 | Esperanza de vida en a√±os.                                                                                                              | Num√©rica             |
| **Adult Mortality**                 | Tasa de mortalidad adulta (probabilidad de morir entre los 15 y 60 a√±os por cada 1000 personas).                                       | Num√©rica             |
| **infant deaths**                   | N√∫mero de muertes infantiles por cada 1000 personas.                                                                                    | Num√©rica             |
| **Alcohol**                         | Consumo per c√°pita de alcohol (en litros de alcohol puro) para personas de 15 a√±os o m√°s.                                              | Num√©rica             |
| **percentage expenditure**          | Gasto en salud como porcentaje del PIB per c√°pita.                                                                                      | Num√©rica             |
| **Hepatitis B**                     | Cobertura de inmunizaci√≥n contra la Hepatitis B en ni√±os de 1 a√±o (%).                                                                  | Num√©rica             |
| **Measles**                         | N√∫mero de casos reportados de sarampi√≥n por cada 1000 personas.                                                                         | Num√©rica             |
| **BMI**                              | √çndice de masa corporal promedio de la poblaci√≥n.                                                                                       | Num√©rica             |
| **under-five deaths**               | N√∫mero de muertes en menores de cinco a√±os por cada 1000 personas.                                                                      | Num√©rica             |
| **Polio**                           | Cobertura de inmunizaci√≥n contra la polio en ni√±os de 1 a√±o (%).                                                                        | Num√©rica             |
| **Total expenditure**               | Gasto general del gobierno en salud como porcentaje del gasto total del gobierno.                                                       | Num√©rica             |
| **Diphtheria**                      | Cobertura de inmunizaci√≥n contra la difteria en ni√±os de 1 a√±o (%).                                                                     | Num√©rica             |
| **HIV/AIDS**                        | N√∫mero de muertes relacionadas con el VIH/SIDA por cada 1000 nacimientos vivos.                                                         | Num√©rica             |
| **GDP**                             | Producto Interno Bruto (PIB) per c√°pita en d√≥lares americanos (USD).                                                                    | Num√©rica             |
| **Population**                      | Poblaci√≥n total del pa√≠s.                                                                                                               | Num√©rica             |
| **thinness 1-19 years**             | Prevalencia de delgadez entre adolescentes de 10 a 19 a√±os (%).                                                                         | Num√©rica             |
| **thinness 5-9 years**              | Prevalencia de delgadez entre ni√±os de 5 a 9 a√±os (%).                                                                                  | Num√©rica             |
| **Income composition of resources** | √çndice de desarrollo humano en t√©rminos de composici√≥n de ingresos (de 0 a 1).                                                          | Num√©rica             |
| **Schooling**                       | A√±os promedio de escolarizaci√≥n.                                                                                                        | Num√©rica             |

```python
# Distribuci√≥n Variable Objetivo:
plt.figure(figsize=(16, 6))
sns.histplot(df['Life expectancy '], bins=50)
plt.title("Distribuci√≥n: 'Life expectancy'")
plt.xlabel("Esperanza de vida")
plt.ylabel("Frecuencia");
```

```python
# Outliers Variable Objetivo:
plt.figure(figsize=(16, 6))
sns.boxplot(x=df['Life expectancy '])
plt.title("Outliers 'Life expectancy'");
```

```python
plt.figure(figsize=(16,16))
sns.heatmap(df.corr(numeric_only=True), annot = True, fmt = '.2f', vmin = -1, cmap = 'coolwarm');
```

```python
sns.heatmap(df[['Adult Mortality', 'Alcohol', 'percentage expenditure', ' BMI ', 'Polio', 'Diphtheria ', 'GDP', 'Income composition of resources', 'Schooling', 'Life expectancy ']].corr(), annot= True, vmin=-1, cmap='coolwarm');
```
